```python
class Person: # 강사님 답변
    # 초기화!!
    hp = 0 # 체력 
    mp = 0 # 마나 

    # hp: 기사, 마법사 / mp: 마법사
    def __init__(self, p_hp, p_mp=None):
        self.hp = p_hp
        if p_mp is not None:
            self.mp = p_mp 

    def rest(self):
        self.hp += 10

        if self.mp: # self.mp: 0 / None -> 둘다 아니다.!!! 
            self.mp += 10


import random

class Knight(Person):
    power = 0

    def __init__(self):
        super().__init__(p_hp=500)
        self.power = random.randint(10,50)

        print(f"knight's hp = {self.hp}\nknight's power = {self.power}")

# 상대방에게 공격
    def attack(self, enemy):
        # 상대방 방어 활률적으로.. 
        enemy.defense(self.power)

    # 방어하자!! 제발.... 살려
    def defense(self, power):
        islife = True
        if random.randint(1,10) < 5:
            print("knight defense fail")
            self.hp -= power/2

        if self.hp <= 0:
            print("죽음")
            islife = False 

        return islife


class Wizard(Person):
    knowledge = 0

    def __init__(self):
        self.knowledge = random.randint(10,50)
        super().__init__(p_hp=300, p_mp=self.knowledge)

        print(f"wizard's hp = {self.hp}\nwizar's power = {self.knowledge}")

    # 상대방에게 공격
    def attack(self, enemy):
        # 상대방 방어 활률적으로.. 
        enemy.defense(self.knowledge)

    # 방어하자!! 제발.... 살려
    def defense(self, power):
        islife = True
        if random.randint(1,10) < 5:
            print("wizard defense fail")
            self.hp -= power

        if self.hp <= 0:
            print("죽음")
            islife = False 

        return islife
```

```python
class Game:
    def __init__(self):
        self.knight = Knight() # 기사 입장
        self.wizard = Wizard() # 마법사 입장

    def start_game(self):
        print("게임을 시작하겠습니다.")
        while True:
            if random.randint(1,10) <5:
                print("knight attack")
                self.knight.attack(self.wizard)
                if self.wizard.hp <= 0:
                    print("knight win")
                    break
            else:
                self.wizard.attack(self.knight)
                print("wizard attack")
                if self.knight.hp <= 0:
                    print("wizard win")
                    break
```

```python
a = Game()
```

```python
a.start_game()
```

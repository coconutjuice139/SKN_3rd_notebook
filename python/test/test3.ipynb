```python
x = np.arange(-5,5,0.1)

y= 1*(x**3)+1*(x**2)+1*(x)+1

y_noise = 15* np.random.normal(size = len(x))
y = y+y_noise

plt.scatter(x,y)
```

```python
x1 = x
x2 = np.power(x1,2)
x3 = np.power(x1,3)
print(len(x1))
```

```python
x1_new = np.reshape(x1,(100,1))
x2_new = np.reshape(x2,(100,1))
x3_new = np.reshape(x3,(100,1))
```

```python
x_bias = np.ones((100,1))
```

```python
x_new = np.append(x_bias, x1_new, axis = 1)
x_new = np.append(x_new, x2_new, axis = 1)
x_new = np.append(x_new, x3_new, axis = 1)
```

```python
x_new_transpose = np.transpose(x_new)
```

```python
x_new_transpose_dot_x_new = x_new_transpose.dot(x_new)
```

```python
temp1 = np.linalg.inv(x_new_transpose_dot_x_new)
temp2 = x_new_transpose.dot(y)
```

```python
theta = temp1.dot(temp2)
theta
```

```python
beta_0 = theta[0]
beta_1 = theta[1]
beta_2 = theta[2]
beta_3 = theta[3]
```

```python
plt.scatter(x,y)
plt.plot(x, beta_0 + beta_1*x1, beta_2*x2, beta_3*x3, c="red")
```

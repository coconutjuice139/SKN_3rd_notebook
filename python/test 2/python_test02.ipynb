```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
X = np.arange(1, 100, 0.5)
Y = 10 + 2*np.log(X)

Y_noise = np.random.rand(len(Y))
Y = Y + Y_noise
plt.scatter(X, Y)
```

```python
n = len(X)
x_bias = np.ones((n, 1))

print(X.shape)
print(x_bias.shape)
```

```python
X_log = np.log(X)
x_log = np.reshape(X_log, (-1, 1))
x_new = np.append(x_bias, x_log, axis=1)
```

```python
x_new_transpose = np.transpose(x_new)
```

```python
x_new_transpose_dot_x_new = x_new_transpose.dot(x_new)
temp1 = np.linalg.inv(x_new_transpose_dot_x_new)
```

```python
temp2 = x_new_transpose.dot(Y)
```

```python
theta = temp1.dot(temp2)
```

```python
theta
```

```python
a = theta[0]
b = theta[1]
Y_plot = a + b*np.log(X)
plt.scatter(X,Y)
plt.plot(X,Y_plot,c='r')
```

```python
Accuracy = r2_score(Y,Y_plot)
print(Accuracy)
```

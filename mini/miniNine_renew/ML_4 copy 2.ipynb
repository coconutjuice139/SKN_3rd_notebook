```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns

def get_clf_eval(y_test, pred):
    confusion = confusion_matrix(y_test, pred)
    accuracy = accuracy_score(y_test, pred)
    precision = precision_score(y_test, pred)
    recall = recall_score(y_test, pred)
    f1 = f1_score(y_test, pred)
    
    print('오차 행렬')
    print(confusion)
    
    # 혼동 행렬 시각화
    plt.figure(figsize=(8, 6))
    sns.heatmap(confusion, annot=True, fmt='d', cmap='RdYlGn')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
    
    # 확률로 표시하고 싶다면 아래 코드를 사용하세요
    # sns.heatmap(confusion / np.sum(confusion), annot=True, fmt='.2%', cmap='Blues')
    
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, F1:{3:.4f}'.format(accuracy, precision, recall, f1))
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import make_scorer, f1_score
import numpy as np

# 사용자 정의 평가 함수 (F1 점수)
def custom_eval(estimator, X, y_true):
    y_pred_proba = estimator.predict_proba(X)[:, 1]
    thresholds = [0.3, 0.4, 0.5, 0.6, 0.7]
    f1_scores = []
    for threshold in thresholds:
        y_pred = (y_pred_proba >= threshold).astype(int)
        f1_scores.append(f1_score(y_true, y_pred))
    return max(f1_scores)

# 사용자 정의 평가 함수를 scorer로 변환
custom_scorer = make_scorer(custom_eval, needs_proba=True)
# random쓰면 param 값이 바뀌어서 결과가 계속 바뀌기 때문에 사용하는 걸 추천하지 않음
param_grid = {
    'C': [0.001, 0.002, 0.003, 0.004],
    'penalty': ['l2'],
    'solver': ['liblinear'],
    'max_iter': [250, 260, 270, 280, 290, 291],
    'class_weight': [None, 'balanced'],
    'tol': [0.00002, 0.00003, 0.00004, 0.00005, 0.00006],
    'fit_intercept': [True, False]
}
SEED = 42
grid_search = GridSearchCV(LogisticRegression(random_state=SEED), param_grid, cv=5, scoring=custom_scorer, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

print("최적 파라미터:", grid_search.best_params_)
# print("최고 F1 점수:", grid_search.best_score_)

best_model = grid_search.best_estimator_
```

```python
# 최적의 threshold 찾기
y_pred_proba = best_model.predict_proba(X_test)[:, 1]
thresholds = [0.3, 0.4, 0.5, 0.6, 0.7] # 여기 threshold 조정할 수 있음
f1_scores = []
for threshold in thresholds:
    y_pred = (y_pred_proba >= threshold).astype(int)
    f1_scores.append(f1_score(y_test, y_pred))

best_threshold = thresholds[np.argmax(f1_scores)]
print("최적 threshold:", best_threshold)

# 최적의 threshold를 사용하여 예측
y_pred = (y_pred_proba >= best_threshold).astype(int)

get_clf_eval(y_test, y_pred)
```

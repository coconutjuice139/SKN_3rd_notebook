# 텍스트 생성
- 텍스트 생서은 OpenAI API의 가장 기본적이면서도 완전 강력한 기능입니다.
- GPT-4와 같은 LLM 언어 모델을 사용하여 사용자 프롬프트(요청)에 기반한 텍스트 생성 기능입니다.

```python
from openai import OpenAI # API 통신용 모듈 
```

```python
client = OpenAI()
```

```python
# OpenAI API 통신 
# completions -> 응답
completions = client.chat.completions.create(
	# 요청을 처리할 LLM 
	model='gpt-4o-mini',
	# 요청 
	messages=[
		{
			"role":'system', # 어시스트 
			"content":"You are a helpful assistant." # LLM 모델의 포지션을 나타냄.
		}, 
		{
			"role":'user', # 요청자(사람 또는 서비스를 받는 어플리케이션)
			"content":"""
				두진을 필두로 한 TEAM 한강경찰대는 좌초된 유람선에서 수상한 작업 중인 기석을 쫓다 한강에서 벌어진 수많은 범죄의 끝에 기석과 경인리버크루즈, 
				그리고 한때 가족과도 같았던 백철이 있다는 것을 알게 된다. 
				TEAM 한강경찰대는 범죄의 소용돌이에 휘말린 한강과 시민들을 지키는 이야기를 만들어주세요.
			""" # 요청내용
		}
	]
)
```
